```python
import pandas as pd
import numpy as np
data=pd.read_csv('train.csv')
testdata=pd.read_csv('test.csv')
data
```

```python
data.info()
```

```python
column_drop=["name","ticket","cabin","embarked","boat","body","home.dest"]
dt=data.drop(column_drop,axis=1)
dtest=testdata.drop(column_drop,axis=1)
dt.head(n=5)
```

```python
dt=dt.fillna(dt["age"].mean())
dtest=dtest.fillna(dtest["age"].mean())
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
dt["sex"]=le.fit_transform(dt["sex"])
dtest["sex"]=le.fit_transform(dtest["sex"])
dt.head()
```

```python
ic = ['pclass',"sex","age","sibsp","parch","fare"]
oc = ["survived"]
x = dt[ic]
y = dt[oc]
from sklearn.tree import DecisionTreeClassifier
tree = DecisionTreeClassifier(criterion='entropy',max_depth=5)
tree.fit(x,y)
s=tree.predict(dtest[ic])
s
```

```python
#out = pd.DataFrame(s)
#df_out['PassengerId'] = dft['PassengerId']
output = pd.DataFrame({"Id":np.arange(300),"survived":s})
output.to_csv('output.csv', sep=',', index=False)
output
```

```python
tree.score(dtest[ic],s)
```
